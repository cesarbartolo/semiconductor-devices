```python
import APD_analysis as apd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=apd.import_data('./apd_power_40V_calmar.csv')
```

```python
power10=.1e-3 #10mW
power12=.063096e-3 #12mW
power14=.039811e-3 #14mW
power16=.025119e-3 #16mW
power18=.015849e-3 #18mW
power20=.01e-3 #20mW
```

```python
EQE10=apd.EQE(data['10dbm']-data['dark'],power10,850e-9)
```

```python
EQE12=apd.EQE(data['12dbm']-data['dark'],power12,850e-9)
```

```python
EQE14=apd.EQE(data['14dbm']-data['dark'],power14,850e-9)
```

```python
EQE16=apd.EQE(data['16dbm']-data['dark'],power16,850e-9)
```

```python
EQE18=apd.EQE(data['18dbm']-data['dark'],power18,850e-9)
```

```python
EQE20=apd.EQE((data['20dbm']-data['dark']),power20,850e-9)
```

```python
EQE20=apd.EQE(data['20dbm']-data['dark'],power20,850e-9)
```

```python
EQE_power=pd.Series([EQE10[-10],EQE12[-10],EQE14[-10],EQE16[-10],EQE18[-10],EQE20[-10]])
```

```python
power=pd.Series([-10,-12,-14,-16,-18,-20])
```

```python
EQE_withPower=pd.DataFrame({
                            'power':power,
                            'EQE':EQE_power    
})#, columns=['power', 'EQE'])
```

```python
plt.scatter(EQE_withPower['power'],EQE_withPower['EQE']*100, s=50)
plt.xlabel('Input power (dBm)')
plt.ylabel('Quantum efficiency (%)')
```
